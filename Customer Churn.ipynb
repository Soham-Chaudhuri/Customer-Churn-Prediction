```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=df[df.TotalCharges!=' ']
```

```python
df.TotalCharges=pd.to_numeric(df.TotalCharges)
df.head()
```

```python
df.dtypes
```

```python
tenure_churn_yes=df[df.Churn=='Yes'].tenure
tenure_churn_no=df[df.Churn=='No'].tenure

plt.xlabel('Tenure')
plt.ylabel('Number of Customers')
plt.title('Customer Churn Prediction Visualization')
plt.hist([tenure_churn_no,tenure_churn_yes],color=['red','green'],label=['Churn=No','Churn=Yes'])
plt.legend()
```

```python
mcharges_churn_yes=df[df.Churn=='Yes'].MonthlyCharges
mcharges_churn_no=df[df.Churn=='No'].MonthlyCharges

plt.xlabel('Monthly Charges')
plt.ylabel('Number of Customers')
plt.title('Customer Churn Prediction Visualization')
plt.hist([mcharges_churn_no,mcharges_churn_yes],color=['red','green'],label=['Churn=No','Churn=Yes'])
plt.legend()
```

```python
def print_unique_col_values(df):
    for column in df:
        if(df[column].dtypes=='object'):
            print(f'{column}:{df[column].unique()}')
```

```python
print_unique_col_values(df)
```

```python
df.replace('No phone service','No',inplace=True)
df.replace('No internet service','No',inplace=True)
```

```python
print_unique_col_values(df)
```

```python
for columns in df:
    if(columns!='InternetService'):
        df.replace({'Yes':1,'No':0},inplace=True)


df.head()
```

```python
print_unique_col_values(df)
```

```python
df.dtypes
```

```python
df.replace({'Male':0,'Female':1},inplace=True)
```

```python
df=pd.get_dummies(data=df,columns=['InternetService','Contract','PaymentMethod'])
```

```python
df.shape
```

```python
df.dtypes
```

```python
from sklearn import preprocessing
```

```python
scaler=preprocessing.MinMaxScaler()
```

```python
df[['tenure','MonthlyCharges','TotalCharges']]=scaler.fit_transform(df[['tenure','MonthlyCharges','TotalCharges']])
```

```python
df.drop('customerID',axis='columns',inplace=True)
df.head()
```

```python
x=df.drop('Churn',axis='columns')
y=df['Churn']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=25)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
import tensorflow as tf
from tensorflow import keras

model=keras.Sequential([
    keras.layers.Dense(200,input_shape=(26,),activation= 'relu'),
    keras.layers.Dense(1,activation= 'sigmoid')
])
model.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=['accuracy']
)

model.fit(x_train,y_train,epochs=200)
```

```python
model.evaluate(x_test,y_test)
```

```python
yp=model.predict(x_test)
yp
```

```python
y_pred=[]
for element in yp:
    if element>0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
print(classification_report(y_test,y_pred))
```

```python
import seaborn as sn
cm=tf.math.confusion_matrix(labels=y_test,predictions=y_pred)
plt.figure(figsize=(10,7))
sn.heatmap(cm,annot=True,fmt='d')
plt.xlabel('predicted')
plt.ylabel('truth')
```
